```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data manually recreated based on the screenshot
data = {
    "Source": ["Ag-Cow", "Ag-Mix", "Ag/Waste-A", "Ag/Waste-B", "Waste-Mix"],
    "CH4": [54.9, 53, 57.7, 56.9, 57.2],
    "CO2": [39, 43.3, 40.8, 41.3, 42.3],
    "N2": [5, 3.2, 1.2, 1.3, 0],
    "O2": [0.9, 0.4, 0.3, 0.5, 0.2],
    "H2 Production": [1.86532, 1.34572, 1.97636, 1.93106, 1.94569]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Create a graph for each source
for source in df['Source']:
    source_data = df[df['Source'] == source]
    
    # Plot H2 production against each composition
    plt.figure(figsize=(8, 6))
    sns.barplot(x=["CH4", "CO2", "N2", "O2"], 
                y=source_data[["CH4", "CO2", "N2", "O2"]].values[0], 
                palette="viridis")
    
    plt.title(f'H2 Production vs Composition for {source}', fontsize=14)
    plt.ylabel('Composition (%)', fontsize=12)
    plt.xlabel('Components', fontsize=12)
    plt.ylim(0, 100)
    
    # Add H2 production as annotation
    plt.annotate(f"H2 Production: {source_data['H2 Production'].values[0]:.2f}", 
                 xy=(2, 50), ha='center', fontsize=12, color='red', fontweight='bold')
    
    plt.show()
```
